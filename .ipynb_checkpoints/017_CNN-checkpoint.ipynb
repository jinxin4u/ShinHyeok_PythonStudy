```python


%matplotlib inline
import scipy as sp
import scipy.misc
import numpy as np
import matplotlib.pylab as plt
import scipy.ndimage # scikit_image

# 흑백이미지를 반전시켜 
img = 255 - sp.misc.face(gray=True).astype(float) # 이미지, 영상
# 경계선 검출 edge detection filter
k = np.zeros((2,2))
k[:,0] = 1; k[:,1] = -1

# convolve는 이미지로부터 필터, 정보통신 무선신호 복원 등
# 통신(노이즈검출), CV(시각검출) 등에 활용됨.
img2 = np.maximum(0, sp.ndimage.filters.convolve(img,k))

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.grid(False)
plt.subplot(122)
plt.imshow(img2)
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
import tensorflow as tf

# 1장, height, width, channels # 차수일치
china = load_sample_image("china.jpg") # 자동 4차원 0~1, 0~255
# 이미지 정규화 : 컬러값을 255로 나눠 0-1로 만들어..
flower = load_sample_image('flower.jpg')
dataset = np.array([china, flower], dtype=np.float32)
# 장수, 높이, 넓이, 컬러수
batch_size, height, width, channels = dataset.shape
print(batch_size, height, width, channels) # 2(장) 427(행; 높이) 640(열; 가로) 3(RGB)

filters =np.zeros( shape=(7, 7, channels, 2), dtype=np.float32)
filters[:, 3, :, 0] = 1
filters[3, :, :, 1] = 1

# 주입변수 427, 640, 3
# 이미지 포맷은 장수,높이,넓이,채널 순으로 고정되어 있다. 맞춰서 넣어줘야한다.
X = tf.placeholder(tf.float32, shape=(None, height, width, channels))

# 427 640
# valid : 421 x 634
# same : 427 x 640

# convolution = tf.nn.conv2d(X, filters, strides=[1,2,2,1], padding="SAME") # same이니까 들어오는 사이즈와 똑같이 나와야한다.
convolution = tf.nn.conv2d(X, filters, strides=[1,1,1,1], padding="SAME") # same이니까 들어오는 사이즈와 똑같이 나와야한다.

# 사이즈 감소
max_pool = tf.nn.max_pool(convolution, ksize=[1,2,2,1], # ksize 2x2를 1로
                          strides=[1,2,2,1], 
                          padding="SAME")

#### 상기는 Graph(node + edge) 정의 ####


with tf.Session() as sess:
    output = sess.run(max_pool, feed_dict = {X:dataset})
# 장 행 열 필터
plt.imshow(output[0, :, :, 1], cmap="gray")
plt.show()
plt.imshow(output[1, :, :, 1], cmap="gray")
plt.show()
plt.imshow(output[1, :, :, 0], cmap="gray")
plt.show()
type(output)
print(output.shape)
```

```python
# 채널분리

import warnings
warnings.filterwarnings(action="ignore")

%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as p
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("/tmp/data/", one_hot = True)

# 5장의 이미지를 출력해보자
# first_image = mnist.test.images[0] #784 한줄로되어있음.
# pixels = first_image.reshape((28, 28))
# plt.imshow(pixels, cmap='gray')
# plt.show()
for i in range(5):
    images = mnist.test.images[i]
    pixels = images.reshape((28,28))
    plt.imshow(pixels, cmap = 'gray')
    plt.show()
```

```python
learning_rate = 0.001 # 학습률 optimizer 사용 할 거
num_steps = 500 # epoch
batch_size = 128 # mini batch size
display_step = 10 # 시각화

# 변수의 갯수가 784라고 말할 수 있다.
num_input = 784 # 이미지 28x28 >> fully-connected : 한줄로 데이터를 연결해놓겠다는 소리
num_classes = 10 # one - hot - encoding
dropout = 0.75 # 과적합 방지를 위해 임의적으로 25%는 계산하지 않음.
# 계산 대상은 랜덤으로 진행. 선택할 수 없음.

# 이미지가 한장씩 들어옴. 784개의 픽셀데이터..
X = tf.placeholder(tf.float32, [None, num_input], name='input') # 훈련 128x784
Y = tf.placeholder(tf.float32, [None, num_classes], name='output') # 종속변수 128x10
keep_prob = tf.placeholder(tf.float32, name='keep') # 드랍아웃
```

```python
# 1번 이상 반복되면 함수화하는 게 좋다.
# activation function은 마지막 한번 호출해서 회귀냐 분류냐를 결정
# 멀티 레이어에서 xor문제해결은 activation function을 이용해서 비선형으로 해결해야
# CNN은 깊을 수록 결과가 좋다

def conv2d(x, W, b, strides=1):
    # 가중치 : filter
    x = tf.nn.conv2d(x, W, strides=[1, strides, strides, 1], padding='SAME')
    x = tf.nn.bias_add(x,b)
    return tf.nn.relu(x) # 비선형으로 만들기 위해 매핑해줌

def maxpool2d(x, k=2):
    return tf.nn.max_pool(x, ksize=[1,k,k,1], # 풀링방법은 max, avg 
                         strides=[1,k,k,1], padding='SAME')
```

```python
def conv_net(x, weights, biases, dropout):#128x784 => 28x28 

    x = tf.reshape(x, shape= [-1, 28, 28, 1]) # 원래 이미지의 형태
    # 128x28x28x1 하나의 이미지로부터 필터가 32장 왜 32번? 각기 다른 특징
    # 
    # 예) 100x100 데이터로부터 5개 특징을 뽑아내겠다하면 
    #     100x5로 => 1차원 벡터 내적
    #
    
    conv1 = conv2d(x, weights['wc1'], biases['bc1']) # 128x28x28x32
    conv1 = maxpool2d(conv1, k=2) # 128x 14x14x32 (특징)
    
    # 한번 특징 뽑아냈으니 풀링해도 특징을 뽑아내는데 이상이 없다.
    # 다음 계산량을 줄이기 위해
    
    conv2 = conv2d(conv1, weights['wc2'], biases['bc2']) # 128x14x14x64
    conv2 = maxpool2d(conv2, k=2) # 128x7x7x64
    
    # CNN의 역할 : 특징 추출 [주변을 고려한 특징 추출] => FFNN : feed forward
    fc1 = tf.reshape(conv2, # 7*7*64, 1024 # 풀리 컨넥티드
                     [-1, weights['wd1'].get_shape().as_list()[0]]) # 장수빼고 7x7x64곱함. 일렬로(1차원으로) 만듬
    # 7*7*64 , 1024 => 128 x 1024   가중치와의 곱으로
    fc1 = tf.add(tf.matmul(fc1, weights['wd1']), biases['bd1'])
    fc1 = tf.nn.relu(fc1)
    
    fc1 = tf.nn.dropout(fc1, dropout) # 25% 는 계산에서 드랍아웃, 드랍아웃도 하이퍼 파라미터
    
    # 128  x 1024, 1024x10 => 128x10
    out = tf.add(tf.matmul(fc1, weights['out']), biases['out'])
    return out

# 하이퍼 파라미터 : 모델이 돌아가기전 정해지는 파라미터[사람이 정하는 거 아님]
# 드랍아웃 : 과적합을 피하기 위해 계산의 일부를 제거
# 이미지 사이즈 결정 -> 필터 사이즈 결정
```

```python
weights = {
    'wc1' : tf.Variable(tf.random_normal([5,5,1,32])),
    'wc2' : tf.Variable(tf.random_normal([5,5,32,64])),
    'wd1' : tf.Variable(tf.random_normal([7*7*64, 1024])),
    'out' : tf.Variable(tf.random_normal([1024, num_classes]))
}

biases = {
    'bc1' : tf.Variable(tf.random_normal([32])),
    'bc2' : tf.Variable(tf.random_normal([64])),
    'bd1' : tf.Variable(tf.random_normal([1024])),
    'out' : tf.Variable(tf.random_normal([num_classes])),
}

# 바이어스 차수는 출력 차수에 따라...
```

```python
logits = conv_net(X, weights, biases, keep_prob)# 128x10로 나옴
prediction = tf.nn.softmax(logits)

loss_op = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits_v2(
    logits=logits, labels=Y))

optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate)
train_op = optimizer.minimize(loss_op)

correct_pred = tf.equal(tf.argmax(prediction, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32), name='accuracy')

tf.add_to_collection("accuracy", accuracy)
tf.add_to_collection("Xi", X)
tf.add_to_collection("Yi", Y)
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for step in range(1, num_steps+1):
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        sess.run(train_op, feed_dict={X:batch_x, Y:batch_y, keep_prob: dropout})
        
        if step % display_step == 0 or step == 1:
            loss, acc = sess.run([loss_op, accuracy], feed_dict={X:batch_x, Y:batch_y, keep_prob:1.0})
            print("Step " + str(step) + ", Minibatch Loss = "+ "{:.4f}".format(loss)
                  + ", Training Accuracy = " + "{:.3f}".format(acc))
    print("Testing Accuracy : ", sess.run(accuracy, feed_dict={X: mnist.test.images[:256],
                                                              Y: mnist.test.labels[:256],
                                                              keep_prob : 1.0}))
```

# 문제 : 테스트 데이터 100번부터 20개의 데이터를 예측해보시오
- 예측된 데이터를 출력해보시오

```python
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(init)
    for step in range(1, num_steps+1) :
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        sess.run(train_op, feed_dict={X: batch_x, Y:batch_y, keep_prob : dropout})
        if step % display_step == 0 or step ==1 :
            loss, acc = sess.run([loss_op, accuracy], feed_dict = {X : batch_x, Y : batch_y, keep_prob : 1.0})
            print('Step' + str(step) + 'Minibatch Loss=' + "{:.4f}".format(loss) + ", Training Accuracy = " + "{:.3f}".format(acc))
    save_path = saver.save(sess, 'data/cnn_model.ckpt')
    print("Testing Accuracy : ", sess.run(accuracy, feed_dict = {X:mnist.test.images[:256],
                                                                Y:mnist.test.labels[:256], keep_prob:1.0}))
    print("예측 : ", sess.run(tf.argmax(prediction, 1), feed_dict={X:mnist.test.images[100:120],
                                                                Y:mnist.test.labels[100:120], keep_prob:1.0}))
        
```
