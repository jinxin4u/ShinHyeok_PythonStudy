```python
df = pd.DataFrame({'value' : np.random.randint(0, 100, 20)}) # 0~100까지 20개

print(df,'\n')

labels = ["{} - {}".format(i, i+9) for i in range(0,100,10)] # 0~100까지 10개 단위로

print(labels,'\n')

# cut method 레퍼런스도 봐야겠다
df['group'] = pd.cut(df.value, range(0, 105, 10), right=False, labels = labels)
# right=false 면 오른쪽 끝수는 포함시키지 않겠다는 의미
# 그래서 0-10 이 아니라 0-9가 들어옴.
# 문자열 라벨을 이용해 범주화

#  범주화할 범위값을 param 2에 주고  param 4에 라벨

print(df,'\n')
```

```python
from pandas import Series, DataFrame
# dict 키는 중복이 안됨
# but DataFrame은 dict의 기능에다가 , 중복을 허용하고, 순서도 보장함.

# merge
df1 = DataFrame({'key' : ['b','b','a','c','a','a','b'],
                 'data1' : range(7)})
print(df1,'\n')

df2 = DataFrame({'key' : ['a','b','d'], 'data2' : range(3)})
print(df2,'\n')

print(pd.merge(df1,df2, how='inner'),'\n') # 양측 다 있는 것만
print(pd.merge(df1,df2, on='key'),'\n') # 키값이 일치하는 것을
print(pd.merge(df1,df2, left_on='key', right_on='key'),'\n') # 만약 키 이름을 다른 데이터의 경우에는 각각 명세해줘야 한다. 
print(pd.merge(df1,df2, how='outer'),'\n') # 양측 데이터 모두(일치하지 않는 것도 다)
```

```python
# 다음 데이터를 18~25 ,25~35, 35~60, 60~100 사이의 값으로 범주화 하시오 
# 범주별로 도수분포표를 작성하시오 

ages = [20,22,25,27,21,24,37,31,61,45,41,32]
bins = [18,25,35,60,100]
pd.cut(ages,bins )
```

```python
def cut(array, bins, labels, closed='right'):
    _bins = pd.IntervalIndex.from_tuples(bins, closed=closed)

    x = pd.cut(array, _bins)
    x.categories = labels
    return x

 

array = [3.5, 1, 0.5, 3]
bins = [(0,1), (1,2), (3,4)]
labels = ['first', 'second', 'third']

df = pd.DataFrame({
    'value': array,
    'category': cut(array, bins, labels, closed='right')
})
df 
```

```python

문제1) 첫번째 컬럼의 이름을 ID-area로 지정하시오.
문제2) regions의 값들을 중복하지 않고 이상한 숫자를 제거하시오.   # 확인 방법은?
문제3) 처음 컬럼(ID_area_)에 들어온 이상한 숫자를 제거하시오.

문제4) 산성관련 성분인 'palmitic', 'palmitoleic', 'stearic', 'oleic', 'linolenic',
        'linolenic', 'arachidic', 'eicosenoic'의 컬럼만 추려서
        별도의 sub 데이터 프레임(변수이름= dfsub)을 생성하시오.
    
문제5) dfsub의 데이터를 모두 100으로 나누어 소수점으로 나타내시오.
문제6) palmitic산과 linolenic산의 분포도를 시각화하시오.
문제7) groupby를 활용하여 region을 기준으로 묶어서 region_groupby 객체로 생성
문제8) region_groupby에 describe() 메소드 적용하여 출력하시오.
```

```python
#1) 문제1) 첫번째 컬럼의 이름을 ID-area로 지정하시오.

# 실제 값을 변경할 때는 inplace = True
olive.rename(columns={olive.columns[0]:'ID_area'}, inplace = True)
olive.columns# 컬럼만 확인
olive.head(5)
```

```python
#2) 문제2) regions의 값들을 중복하지 않고 이상한 숫자를 제거하시오.   # 확인 방법은?

olive.region.unique()

```

```python
olive.area.unique()
```

```python
# 참고 : crosstab: 교차분석표
pd.crosstab(olive.area, olive.region)
```

```python
olive.head()
```

```python
#3)  문자열 처리  또는 "."
#문제3) 처음 컬럼(ID_area_)에 들어온 이상한 숫자를 제거하시오.

olive["ID_area"]= olive["ID_area"].apply(lambda x: x.split('.')[1]) # 람다 함수를 통해 하나만 처리
```

```python
olive.head()
```

```python
#4) 문제4) 산성관련 성분인 'palmitic', 'palmitoleic', 'stearic', 'oleic', 'linolenic',
# 'linolenic', 'arachidic', 'eicosenoic'의 컬럼만 추려서
#  별도의 sub 데이터 프레임(변수이름= olivesub)을 생성하시오.

acidlist=['palmitic', 'palmitoleic', 'stearic', 'oleic', 'linolenic',
        'linolenic', 'arachidic', 'eicosenoic']
olivesub= olive[acidlist]
olivesub.head()
```

```python
#5) 문제5) dfsub의 데이터를 모두 100으로 나누어 소수점으로 나타내시오.


olivesub=olive[acidlist].apply(lambda x: x/100.0)

olive[acidlist]=olivesub  # 원본데이터에 다시 집어넣음
```

```python
olive.head()
```

```python
#6) 문제6) palmitic산과 linolenic산의 분포도를 시각화하시오.
%matplotlib inline
import matplotlib.pyplot as plt

fig=plt.figure()
plt.scatter(olive["palmitic"], olive["linolenic"])
axis= fig.gca()
axis.set_title('linolenic vs palmitic')
axis.set_xlabel('palmitic')
axis.set_ylabel('linolenic')
```

```python
#7) 문제7) groupby를 활용하여 region을 기준으로 묶어서 region_groupby 객체로 생성

region_groupby= olive.groupby(['region'])
# 객체라서 따로 내용을 봐야함.(요약함수, 집계함수 써야함.)

region_groupby.apply(np.mean)   # 시리즈는 np 밑에 있다.  포인터(주소값)

# 여러개 적용은 aggFunc


```
